```python
# Import your libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Import the dataset
station = pd.read_csv('station.csv')
print(station.head())
```

```python
# Create a line plot showing the temperature vs the Annual Mean Temperature (metANN)

plt.figure()  # Create a new figure
sns.lineplot(data=station, x='YEAR', y='metANN')

plt.title("Annual Mean Temperature Over Time")
plt.ylabel("Temperature (°C)")

plt.show()
```

```python
# This plot doesn't look quite right! It turns out the problem is that the dataset
# contains a placeholder value of 999.90 for missing data. Let's clean that up.

station['metANN Clean'] = station['metANN'].replace(999.90, np.nan) # This code replaces 999.90 with NaN

plt.figure()

sns.lineplot(data=station, x='YEAR', y='metANN Clean')

plt.title("Annual Mean Temperature Over Time")
plt.ylabel("Temperature (°C)")

plt.show()

# Include your final plot in your homework submission. 
# Don't take a screenshot, save the plot as an image file and then you can copy and paste
# it into your homework submission.
```
