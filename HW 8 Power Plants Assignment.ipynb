```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
power = pd.read_csv('data/global_power_plant_database.csv')
```

```python
# This code will create a data set of just gas power plants that was used to make the plot at the top of this notebook. 
gas = power[power['fuel1'] == 'Gas'] # Filter on Gas
gas_by_country = gas.groupby('country_long').size().reset_index(name='count') 
gas_filtered = gas_by_country[gas_by_country['count'] > 50] 


coal = power[power['fuel1'] == 'Coal']  
coal_by_country = coal.groupby('country_long').size().reset_index(name='count')  
coal_filtered = coal_by_country[coal_by_country['count'] > 50]  


sns.barplot(data=coal_filtered, x='country_long', y='count')
plt.title('Number of Coal Power Plants by Country (Countries with >50 Plants)')
plt.xlabel('Country')
plt.ylabel('Number of Coal Power Plants')
plt.xticks(rotation=45)
plt.show()
```
