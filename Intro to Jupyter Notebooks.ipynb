# Introduction to Jupyter Notebooks

Welcome! Jupyter notebooks are interactive documents that combine:
- **Code** that you can run
- **Text** (and graphics) for explanations and documentation
- **Outputs** like numbers, graphs, and tables

They're widely used in data science, research, and teaching because you can write code, see results immediately, and explain what you're doing all in one place.

## Cell Types

Jupyter notebooks have two main types of cells:

1. **Markdown cells** - Like this one! Used for text, headings, lists, and explanations. Double click on this text to edit it. (Hit the check mark when done)
2. **Code cells** - Contain Python code that can be executed

You can tell the difference by looking at the left side:
- Markdown cells have no brackets
- Code cells have `[ ]` or `[1]`, `[2]`, etc. after they run

## Understanding the Kernel

The **kernel** is the Python engine running in the background that:
- Executes your code
- Stores all your variables in memory
- Keeps track of what you've run

### Important Kernel Commands:

**Restart:**
- Clears ALL variables from memory
- Resets the execution count to [1]
- Like starting fresh, useful if you have imported some data and changed it by mistake. 

**Run All:**
- Runs all cells in your notebook from top to bottom
- Great way to make sure your notebook works from start to finish

You can find these options in the toolbar at the top of the notebook.

## Best Practices

Before you finish working on a notebook, follow these best practices:

### 1. Run Cells in Order (Top to Bottom)
- Cells can technically be run in any order
- But this can create confusing errors
- Always develop your notebook working from top to bottom

### 2. Use Markdown Cells to Document
- Explain what your code does
- Add section headers
- Include your thought process
- Future you (and others) will thank you!

### 3. Restart & Run All Before Submitting
- Ensures your notebook works from scratch
- Catches errors from running cells out of order
- Makes your work reproducible

### 4. Save Frequently
- Notebooks should auto-save, but double check that feature is on in File->Auto Save

## Committing Your Work to GitHub

GitHub is a cloud based repository for storing code. It keeps track of all of the changes you have made to your notebooks over time, a nice feature if you make a mistake and need to go back to a previous version. 

### Steps for GitHub Codespaces:

1. **Open Source Control** - Click the Source Control icon in the left sidebar (looks like a branch icon) or press Ctrl/Cmd + Shift + G

1. **Write a commit message** - In the text box at the top, write a brief message describing your changes (e.g., "Completed Jupyter intro notebook")

1. **Commit** - Click the **Commit** button (checkmark icon)
   - **First time only:** VS Code will ask if you want to automatically stage all changes - click **Always** (this makes future commits easier!)

1. **Push to GitHub** - Click the **Sync Changes** button to push your commits to GitHub (this stores online in your GitHub repository)

Your work is now saved to GitHub! 🎉
